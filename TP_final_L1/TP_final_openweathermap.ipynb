```python
from api_key_utils import api_key_utils
from get_weather_data import get_weather_data
from process_weather_data import process_weather_data
from normalize_data import normalize_data


def main():
    # Obtener la clave API
    api_key = api_key_utils()

    # Definir una lista de coordenadas de ciudades que deseas consultar
    coords_list = [
        (-27.46056, -58.983891),   # Resistencia, AR
        (51.512791, -0.09184),     # City of London, GB
        (40.714272, -74.005966),   # New York, US
        (37.90448, -4.77768),      # Cordoba, ES
        (25.025881, 121.651611),   # Taipei, TW
        (-34.599998, -58.450001),  # Ciudad AutÃ³noma de Buenos Aires, AR
        (19.428471, -99.127663),   # Mexico City, MX
        (53.355122, -6.24922),     # Dublin City, IE (Ireland)
        (41.694111, 44.833679),    # Tbilisi, GE (Georgia)
        (4.60971, -74.081749),     # BogotÃ¡, CO
        (35.689499, 139.691711)    # Tokyo, JP
    ]

    # Llamar a la funciÃ³n get_weather_data() con la lista de coordenadas y la clave de la API
    weather_data_list = get_weather_data(coords_list, api_key)
    # print("JSON completo:", weather_data_list)

    if weather_data_list:
        # call the function
        processed_data = process_weather_data(weather_data_list)
        print("JSON filtrado:\n", processed_data)
        print("--------------------------------")

        # call the method that creates the csv
        normalize_data(processed_data)


if __name__ == '__main__':
    main()
```
